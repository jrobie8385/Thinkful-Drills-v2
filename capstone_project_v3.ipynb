```python
all_data.groupby("emp_length").loan_amnt.sum().plot(kind="bar", figsize=(10,5))
plt.ylabel("Loan Amount")
plt.xlabel("Length of Employment")

message = "Key Takeaway: This chart indicates that smaller total laons are given out to customers as their work history grows.  The largest total laons are made to people with 10+ years of work experience because it is a \"catch all\" category that includes the largest number of people."
print(message)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(all_data.groupby("emp_length").loan_amnt.sum(), kind = "line", color = "green")
plt.ylabel("Loan Amount")
plt.xlabel("Length of Employment")

plt.subplot(1,2,2)
plt.plot(all_data.groupby("emp_length").value_counts, kind = "line", color = "red")
plt.ylabel("Loan Amount")
plt.xlabel("Length of Employment")

message = "Key Takeaway: This chart indicates that a fewer number of laons are given out to customers as their work history grows.  The largest total laons are made to people with 10+ years of work experience because it is a \"catch all\" category that includes the largest number of people."
print(message)
```
